```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('E:\python\olympics.csv')
```

```python
df.head()
```

```python
rows , cols = df.shape
print("Total Rows are "+str(rows))
print("Total Columns are "+str(cols))
```

```python
plt.figure(figsize=(10, 5))
x = ["USA", "CHN", "JPN", "AUS", "FRA"]
y = [126, 91, 45, 53, 64]
plt.bar(x, y, label='Общее кол-во медалей')
plt.xlabel('Страна')
plt.ylabel('Кол-во медалей')
plt.legend()
plt.show()
```

```python
df1=df.groupby("country_code").agg({"total":sum}).reset_index().sort_values(by='total',ascending=False)
plt.figure(figsize=(15,5))
plt.xticks(rotation=20)
plt.title("Страны и их кол-во медалей (общее)")
sns.barplot(x='country_code',y='total',data=df1.head(20), palette='viridis')
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['gold'], bins=10, kde=True, color='gold')
plt.title('Распределение золотых медалей')
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(df['silver'], bins=10, kde=True, color='silver')
plt.title('Распределение серебрянных медалей')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
x = df[('total')]
y = df[('population')]
plt.bar(x, y)
plt.xlabel('Кол-во медалей')
plt.ylabel('Население (в млн.)')
plt.title('Отношение населения страны и медалей')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
x = ['USA', 'CHN', 'JPN', 'AUS', 'FRA']
y = [334.9, 1410.7, 124.5, 26.6, 68.2]
plt.bar(x, y)
plt.xlabel('Страны')
plt.ylabel('Население (в млн.)')
plt.title('Население стран')
plt.show()
```

```python
vals = [81695.19, 12614.06, 33834.39, 64711.77, 44460.82]
labels = ["USA", "CHN", "JPN", "AUS", "FRA"]

plt.pie(vals, labels=labels, autopct='%1.1f%%')
plt.title("Распределение ВВП стран")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
x = df[('region')]
y = df[('total')]
plt.bar(x, y)
plt.xlabel('Регион')
plt.ylabel('Кол-во медалей')
plt.title('Количетсво медалей у Регионов')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='region', y='total', hue='region', size='total', sizes=(20, 200), alpha=0.7)
#Соотношение регионов и медалей
```

```python
top_countries = df.loc[df.groupby('region')['total'].idxmax()]
top_countries[['region', 'country', 'total']]
#Страны получившие больше всего медалей в своем регионе
```
